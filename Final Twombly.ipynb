I had not analyzed why the 2005 work “Untitled” of Cy Twombly’s Bacchus series resonates so deeply with me until undertaking this exercise. Looking into the history of Twombly to contextualize his work made it feel all the more appropriate to try and replicate and iterate on “Untitled” to communicate my own observations and frustrations in this moment.

Professor Katharina Schmidt links Twombly’s style to his work as an army cryptographer characterizing it as, “purely graphic marks… a kind of meta-script in which abbreviated signs, hatchings, loops, numbers and the simplest of pictographs spread throughout the picture plane in a process of incessant movement, repeatedly subverted by erasures”.

Through this ascemic method, Twombly took well known mythological episodes and communicated them through his own language - encoding parallels to contemporary events as a form of social commentary. In searching for analyses of the events Twombly might have alluded to in “Untitled”, the only commentaries I could find fixated on the piece’s brash palate and intimidating stature. However, one analysis by art critic Roberta Smith stood out – “During his [Twombly’s] last decade, he amplified their scale and exuberance, using a wide brush to give this favored motif an impassioned last hurrah”.

Perhaps “Untitled” was unique to Twombly’s repertoire in that it wasn’t a direct critique of a particular societal event, but a communication of a sense of self. I read “Untitled” as a transmission of raw emotion, a directionless and visceral frustration stemming from a sense societal incongruence simultaneously muted and celebrated in bacchanalia. Either that or I am projecting my own feelings of frustration onto this piece – exacerbated by the unnaturalness of life in quarantine.

```python
from IPython.display import SVG, display
def show(page):
    display(SVG(page.svg()))
```

```python
from scipy.stats import truncnorm
def t_normal(a, b, mu, sigma):
    tn = truncnorm((a-mu)/sigma, (b-mu)/sigma, loc=mu, scale=sigma)
    return tn.rvs(1)[0]
```

Generates one form of glyph that is less compact.

```python
def uniform_glyph(n=5):
    pts = []
    for i in range(n):
        pts.append([uniform(-0.5, 0.8), uniform(-0.5, 0.8)])
    rand_pline = Polyline(pts)
    return rand_pline
```

Though this cell generates the glyph we will be working with

```python
def make_char():
    pts = []
    for i in range(int(t_normal(100, 150, 80, 255))):
        x = (choice([0, 0.3, 1]))
        y = (choice([0, 0.5, 1]))
        pts.append([x, y])
    return Polyline(pts)
```

```python
line = make_char()
for i in range(1, 20):
    line += make_char().translate(i*1.2, 0)
line = line.scale(80).translate(00, 20)
d = document(250, 100, 'mm')
page = d.addpage()
line_fig = shape().stroke(rgba(0, 0, 0, 255)).nofill()
curve_fig = shape().stroke(rgba(196, 22, 29, 255)).width(10)
page.place(line_fig.polyline(line))
page.place(curve_fig.path(line.smooth_path()))
show(page)
```

```python
color_pick = widgets.ColorPicker()
display(color_pick)
```

```python
def hex2rgb(s):
    s = s.lstrip("#")
    return [int(s[i:i+2], 16) for i in range(0,5,2)]
hex2rgb(color_pick.value)
```

```python
line = make_char()
for i in range(1, 20):
    line += make_char().translate(i*1.2, 0)
line = line.scale(80).translate(00, 20)
d = document(250, 100, 'mm')
page = d.addpage()

#background
figure = shape().fill(rgba(227, 199, 162, 255)).stroke(rgba(0, 0, 0, 0)).width(0)
rect = figure.rectangle(0, 0, 500, 500)
page.place(rect)

#glyph
line_fig = shape().stroke(rgba(0, 0, 0, 255)).nofill()
curve_fig = shape().stroke(rgb(*hex2rgb(color_pick.value))).width(10)
page.place(line_fig.polyline(line))
page.place(curve_fig.path(line.smooth_path()))
show(page)
```
