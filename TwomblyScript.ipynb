Brush Effect
- I wanted to make the strokes of the lines seem as if they were "brushed" onto the digital canvas
- To do so, I extracted example code from Allison Parrish's "Flat, asemic writing, Bezier curves" python file found here:
    https://mol.decontextualize.com/schedule/

```python
pts = []
for i in range(25):
    pts.append([normal(), normal()])
pline = Polyline(pts)

# draw polyline many times on top of itself
d = document(250, 50, 'mm')
page = d.addpage()
figure = shape().stroke(rgba(196, 22, 29, 225)).width(1)
for i in range(50):
    # add a bit of noise to the coordinates on each loop
    pline.vertices += np.random.randn(2) * 1.5
    lines = figure.polyline(pline.scale(100).translate(100, 0))
    page.place(lines)
show(page)
```
